```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

raw_insurance_df = pd.read_csv('insurance.csv')
print(raw_insurance_df.head())
```

```python
# Finding out data about the columns
print(raw_insurance_df.info())
```

```python
# Calculate summary statistics
print(raw_insurance_df.describe())
```

```python
# checking if there are any nulls in the data
print(raw_insurance_df.isnull().sum())
```

```python
# plot distribution of BMI
sns.boxplot(x='bmi',data=raw_insurance_df)
plt.show()
plt.clf()
```

```python
# histogram for bmi values
sns.histplot(x='bmi',data=raw_insurance_df)
plt.show()
plt.clf()
```

```python
# plot distribution of age
sns.boxplot(x='age',data=raw_insurance_df)
plt.show()
plt.clf()
```

```python
# histogram for age values
sns.histplot(x='age',data=raw_insurance_df)
plt.show()
plt.clf()
```

```python
# plot distribution of charges
sns.boxplot(x='charges',data=raw_insurance_df)
plt.show()
plt.clf()
```

```python
# histogram for charges values
sns.histplot(x='charges',data=raw_insurance_df)
plt.show()
plt.clf()
```

```python
# Create bar chart of smoker vs non smoker
sns.countplot(x='smoker', data=raw_insurance_df)
plt.show()
plt.clf()
```

```python
print(raw_insurance_df.smoker.value_counts())
print(raw_insurance_df.smoker.value_counts(normalize=True))
```

```python
# Show bar chart for regions
sns.countplot(x='region',data=raw_insurance_df)
plt.show()
plt.clf()
```

```python
# let's work with some proportions
print(raw_insurance_df.region.value_counts())
```

```python
print(raw_insurance_df.region.value_counts(normalize=True))
```

```python
# Create bar chart of male vs female
sns.countplot(x='sex', data=raw_insurance_df)
plt.show()
plt.clf()
```

```python
print(raw_insurance_df.sex.value_counts())
print(raw_insurance_df.sex.value_counts(normalize=True))
```

```python
# see charges between male and female
sns.boxplot(data=raw_insurance_df, x='sex',y='charges')
plt.show()
```

```python
# see charges between smoker and non smoker
sns.boxplot(data=raw_insurance_df, x='smoker',y='charges')
plt.show()
```

```python
# see charges between regions
sns.boxplot(data=raw_insurance_df, x='region',y='charges')
plt.show()
```

```python
# relationship between children and bmi
plt.scatter(x=raw_insurance_df.children,y=raw_insurance_df.bmi)
plt.xlabel('Number of Children')
plt.ylabel('BMI value')

plt.show()
```
